# The Teleportation protocol

In this notebook we'll implement the teleportation protocol, to teleport a randomly sampled qubit state from one party to another. That is we start with a state

$$
|\psi\rangle_{A_0} |\Phi_{00}\rangle_{A_1 B} 
$$

where $|\psi\rangle_{A_0}$ is an unknown qubit state and $|\Phi_{00} \rangle_{A_1 B} = (|00\rangle + |11\rangle)/\sqrt{2}$ is an entangled state shared between Alice and Bob.

```python
# Do the necessary imports
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit import IBMQ, transpile, assemble
from qiskit.providers.aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.extensions import UnitaryGate
from qiskit.quantum_info import random_unitary
from qiskit.result import marginal_counts
```

```python
## SETUP
# Protocol uses 3 qubits and 3 classical bits in 3 different registers

qr = QuantumRegister(3, name="q")            # Initialize 3 qubits
crz = ClassicalRegister(1, name="crz")       # Classical bit to check for Z correction
crx = ClassicalRegister(1, name="crx")       # Classical bit to check for X correction
check = ClassicalRegister(1, name="check")   # Classical bit to check we correctly teleported the state
circuit = QuantumCircuit(qr, crz, crx, check)
```

## Step 0 - Initialization

We begin by creating a random state that Alice will teleport

$$
|\psi\rangle_{A_0} = U |0\rangle
$$

where $U$ is a randomly chosen quantum gate. After the protocol has finished, we will use this unitary matrix again to allow us to check that the teleportation was successful. 

```python
## Step 0.1 
U = UnitaryGate(random_unitary(2),label='U')
circuit.append(U,[0])
circuit.barrier()
circuit.draw('mpl', initial_state=1)
```

In order to perform the teleportation we need some entanglement between Alice and Bob. In particular we will use the state 

$$
(|00\rangle + |11\rangle)/\sqrt{2}
$$

shared between Alice and Bob. (In the circuit Alice holds qubits 0 and 1 and Bob holds qubit 2).

```python
## Step 0.2
# We distribute entanglement between Alice and Bob 
circuit.h(1)
circuit.cx(1,2)
circuit.barrier()
circuit.draw('mpl', initial_state=1)
```

## Step 1 - Measurement and communication 

In order to teleport the quantum state Alice will measure the her two qubits in a special basis (known as the Bell-basis). The Bell basis is an orthonormal basis for two qubits made from entangled states 

$$
\begin{aligned}
|\Phi_{00}\rangle &= (|00\rangle + |11\rangle)/\sqrt{2} \\
|\Phi_{01}\rangle &= (|01\rangle + |10\rangle)/\sqrt{2} \\
|\Phi_{10}\rangle &= (|00\rangle - |11\rangle)/\sqrt{2} \\
|\Phi_{11}\rangle &= (|01\rangle - |10\rangle)/\sqrt{2}
\end{aligned}
$$

you can check that this is really an orthonormal basis. 

**Question**: How does the circuit below measuring in this basis?

Alice receives a two bit outcome for the measurement and communicates this to Bob.


```python
## Step 1
# Alice performs the measurement and sends the results to Bob
circuit.cx(0,1)
circuit.h(0)
circuit.measure(0,0)
circuit.measure(1,1)
circuit.barrier()
circuit.draw('mpl', initial_state=1)
```

## Step 2 - Correcting the teleported qubit

Based on the outcome of the measurement Bob's qubit will be transformed into one of four states

$$
\begin{aligned}
|\psi\rangle& \\
X |\psi\rangle& \\
Z |\psi\rangle& \\
XZ|\psi\rangle
\end{aligned}
$$

for outputs 00, 01, 10 and 11 respectively. He can then apply quantum gates to his qubit in order to correct for the different measurement outcomes and recover $|\psi\rangle$. 

```python
## Step 2
# Bob corrects for Alice's measurement outcome
circuit.x(2).c_if(crx,1)    # If the second bit was 1 then need to apply X
circuit.z(2).c_if(crz,1)    # If the first bit was 1 then need to apply Z
circuit.draw('mpl', initial_state=1)
```

Et voila! Assuming no errors occured then Alice has successfully teleported her qubit to Bob :)

# Circuit analysis
## Let's check that the circuit is really working...

Teleportation is especially useful as Alice can teleport **unknown** quantum states to Bob. But if we assume for the moment that we know the state that was sent, we can check that this is indeed the state that Bob received. 

**Question**: Knowing the unitary $U$ such that $|\psi\rangle = U |0\rangle$,  how can we check that Bob received $|\psi\rangle$?


<details>
<summary>SOLUTION (click here)</summary>
If Bob really received the state $U |0\rangle$, he can apply the inverse gate $U^{\dagger}$ to transform his state to $U^{\dagger} U |0\rangle = |0\rangle$. If he then measures in the computational basis he should always receive the output 0. By repeating this procedure many times he can become very confident that the circuit is working correctly!
</details>

```python
# If Alice tell's Bob what unitary she applied
# then Bob can check he received the correct state! 

circuit.barrier()
circuit.append(U.adjoint(),[2])
circuit.measure(2,2)
circuit.draw('mpl', initial_state=1)
```

## Let's simulate the circuit and check it is working correctly!

```python
# Let's simulate the results
sim = AerSimulator()                                   # Choose our simulator backend
compiled_circuit = transpile(circuit, sim)             # Compile the circuit
job = sim.run(compiled_circuit, shots = 1000)          # Choose the number of times to run the circuit and run
result = job.result()                                  # Get the results
counts = result.get_counts(circuit)                    # Counts for all 3 measurements 
marginal_result = marginal_counts(result, [2])         # We only need the counts for the final measurement 
check_counts = marginal_result.get_counts(circuit)     # so we marginalize over the other two measurements 
plot_histogram(check_counts)                           # Plot the counts
```

We see that the result is always the output $0$ so Bob is successfully receiving the state!
